```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount (₹): "))

        if amount < 50:
            print("❌ Recharge amount must be at least ₹50. Try again.")
        elif amount not in valid_plans:
            print("❌ Invalid plan. Valid plans are:", valid_plans)
            print("Please try again.")
        else:
            print(f"✅ Recharge of ₹{amount} successful!")
            break


# Function call
validate_recharge()
```

```python
def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(f"{product}: ⚠️ Reorder Alert (Stock = {quantity})")
        else:
            print(f"{product}: ✅ Stock OK (Stock = {quantity})")


# Sample dictionary of products and stock quantities
inventory = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 30
}

# Function call
check_inventory(inventory)
```

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average = {average:.2f} → Pass"
    else:
        return f"Average = {average:.2f} → Fail"


# Sample input
student_marks = [60, 55, 40, 65, 50]

# Function call
result = process_result(student_marks)
print(result)
```

```python
def calculate_fare():
    BASE_FARE = 50
    PER_KM_RATE = 12

    while True:
        km = float(input("Enter distance travelled (in km): "))
        peak = input("Is it peak hour? (yes/no): ").lower()

        fare = BASE_FARE + (km * PER_KM_RATE)

        if peak == "yes":
            fare += fare * 0.25   # 25% extra during peak hour

        print(f"Total Cab Fare = ₹{fare:.2f}")

        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using the Cab Fare Estimator!")
            break


# Function call
calculate_fare()
```

```python
def check_eligibility(attendance):
    present_days = 0

    for day in attendance:
        if day == 'P':
            present_days += 1

    total_days = len(attendance)
    attendance_percentage = (present_days / total_days) * 100

    if attendance_percentage >= 75:
        return f"Attendance = {attendance_percentage:.2f}% → Eligible"
    else:
        return f"Attendance = {attendance_percentage:.2f}% → Not Eligible"


# Sample attendance list
employee_attendance = ['P', 'A', 'P', 'P', 'A', 'P', 'P', 'P', 'P']

# Function call
result = check_eligibility(employee_attendance)
print(result)
```

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "❌ Weak Password (Minimum length is 8)"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "✅ Strong Password"
    else:
        return "❌ Weak Password (Must contain at least one digit and one special character @#$)"


# Sample test
password = input("Enter password: ")
result = check_password_strength(password)
print(result)
```
